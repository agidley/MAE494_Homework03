```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def f(x, y):
    return (x + 1)**2 + (y - 2)**2

# Define the range of x and y values
x = np.linspace(0, 2, 400)
y = np.linspace(0, 1, 400)

# Create a grid of x and y values
X, Y = np.meshgrid(x, y)

# Compute the function values for each point on the grid
Z = f(X, Y)

# Create a 2D contour plot
plt.figure(figsize=(10, 4))
plt.subplot(121)  # Create a subplot (1 row, 2 columns, 1st plot)
plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour Plot of (x+1)^2 + (y-2)^2')

# Create a 3D surface plot
ax = plt.subplot(122, projection='3d')  # Create a subplot (1 row, 2 columns, 2nd plot)
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y)')
plt.title('3D Surface Plot of (x+1)^2 + (y-2)^2')

# Show the plots
plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return -x

# Define the range of x and y values over a wider domain
x = np.linspace(0, 1.25, 200)
y = np.linspace(0, 1.25, 200)

# Create a grid of x and y values
X, Y = np.meshgrid(x, y)

# Apply the constraints
constraint1 = Y - (1 - X)**3
constraint2 = -Y  # y >= 0

# Create a grid to evaluate the function values
Z = f(X)

# Apply constraints to filter the values of Z
Z[constraint1 > 0] = np.nan
Z[constraint2 > 0] = np.nan

# Create a 2D contour plot
plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour Plot of f(x) = -x with Constraints')

# Add constraint lines
plt.plot(x, (1 - x)**3, 'r--', linewidth=2, label='y - (1 - x)^3 <= 0')
plt.axhline(0, color='g', linestyle='--', linewidth=2, label='y >= 0')

plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def f(x):
    return -x

# Define the range of x and y values
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)

# Create a grid of x and y values
X, Y = np.meshgrid(x, y)

# Apply the constraints
constraint1 = Y - (1 - X)**3
constraint2 = -Y  # y >= 0

# Create a grid to evaluate the function values
Z = f(X)

# Apply constraints to filter the values of Z
Z[constraint1 > 0] = np.nan
Z[constraint2 > 0] = np.nan

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Label the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y)')

# Set axis limits
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

# Display the plot
plt.title('3D Plot of f(x) = -x with Constraints')
plt.show()


```
